## Feature Selection Techniques

1. While collecting data, there might be humar error, duplicate features or bad data.

2. The independent features(many) and dependent feature can also be sometimes related so we can drop some of the independent features

```python
import pandas as pd
df = pd.read_csv('mobile_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

### 1. Univariate Selection

```python
X = df.iloc[:,:-1]
y = df['price_range']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.feature_selection import chi2,SelectKBest
```

#### Apply SelectKBest Algorithm

* Also Called as **InformationK**

```python
X.shape
```

```python
ordered_rank_features = SelectKBest(score_func=chi2, k=20)
ordered_feature = ordered_rank_features.fit(X,y)
```

```python
ordered_feature
```

```python
ordered_feature.scores_
```

```python
dfscores = pd.DataFrame(ordered_feature.scores_, columns=["Score"])
dfcolumns = pd.DataFrame(X.columns)
```

```python
features_rank = pd.concat([dfcolumns, dfscores], axis=1)
```

```python
features_rank.columns = ['Features', 'Score']
features_rank
```

* **Higher the score, more important is the feature.**

#### Taking the Top 10 features

```python
features_rank.nlargest(10,'Score')
```

### 2. Feature Importance

* This technique gives you a score for each feature of your data,the higher the score more relevant and important it is.

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt

model=ExtraTreesClassifier()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
ranked_features=pd.Series(model.feature_importances_,index=X.columns)
ranked_features.nlargest(10).plot(kind='barh')
plt.show()
```

#### Removing the correlated features w.r.t to only Independent Features (**NOT DEPENDENT FEATURES**)

```python
# Threshold is decided by the domain expert or the stake holders.

threshold = 0.5
```

```python
# Find and remove correlated features (As both the features do the same task --> Remove 1 feature.)

def correlation(dataset, threshold):
    
    # Set of all the names of correlated columns (Unique Values only)
    col_corr = set()
    corr_matrix = dataset.corr()
    
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            
            # We are interested in absolute coefficient value.
            if abs(corr_matrix.iloc[i, j]) > threshold:
                
                # Getting the name of feature.
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
correlation(df.iloc[:,:-1],threshold)
```

**Harm in Keeping Correlated Features**

1. X, y --> Independent Features and z --> Dependent Feature.

2. In ML, we need the independent features to be directly dependent or propotinal to the dependent feature.

3. X and y should not be dependent on each other and should be completely independent of each other.

4. If the independent features are dependent on each other, they do the same task and are of no use.

### 4. Information Gain

* **Mutual information (MI)** between two random variables is a non-negative value, which measures the dependency between the variables. 


* It is equal to zero if and only if two random variables are independent, and **higher values mean higher dependency. (Target Variable or Dependent Variable)**

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
mutual_info=mutual_info_classif(X,y)
```

```python
mutual_data=pd.Series(mutual_info,index=X.columns)
mutual_data.sort_values(ascending=False)
```
