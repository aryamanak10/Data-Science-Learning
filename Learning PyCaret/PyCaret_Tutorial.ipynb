## Handling Missing Values using PyCaret

```
from pycaret.datasets import get_data
hepatitis = get_data('hepatitis')
```

```
# Data Preprocessing

from pycaret.classification import *
clf = setup(data=hepatitis, target='Class')
```

### Ignoring Features

```
pokemon = get_data('pokemon')
```

```
clf = setup(data=pokemon, target='Legendary', ignore_features=['#', 'Name'])
```

### Changing Data Types

```
hepatitis = get_data('hepatitis')
```

```
clf = setup(data=hepatitis, target='Class', categorical_features=['AGE'])
```

### One-Hot Encoding

```
clf = setup(data=pokemon, target='Legendary')
```

### Ordinal Encoding (Rank can be given)

```
employee = get_data('employee')
```

```
clf = setup(data=employee, target='left', ordinal_features={'salary': ['low', 'medium', 'high']})
```

```
clf
```

### Cardinal Encoding

```
income = get_data('income')
```

```
income['native-country'].unique()
```

```
clf = setup(data= income, target = 'income >50K', high_cardinality_features = ['native-country'])
```

```
clf
```

### Normalization

```
clf = setup(data=pokemon, target='Legendary', normalize=True)
```

```
clf
```

### Outliers

```
insurance = get_data('insurance')
```

```
from pycaret.regression import *
reg1 = setup(data = insurance, target = 'charges', remove_outliers = True)
```

### Feature Importance

```
diabetes = get_data('diabetes')
```

```
clf1 = setup(data = diabetes, target = 'Class variable', feature_selection = True)
```
